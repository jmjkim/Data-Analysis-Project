```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# IBM's Professional Data Analyst (IBM DA0100 Data Analytics Basics for Everyone)

pd.set_option("display.max_columns", None)

csv = pd.read_csv("Final Assignment.csv")
df = pd.DataFrame(csv)

# Product Category = Designer Clothes
dc = df["Product Category"] == "Designer Clothes"
dc_table = df[dc]

plt.figure(figsize = (12, 8))
plt.ylabel("Designer Clothes")
plt.setp(plt.bar(dc_table["Month-Year of Purchase"], dc_table["Units Sold"]), color = "orange", width = 0.4)

# Product Category = Fitness Gadgets
fg = df["Product Category"] == "Fitness Gadgets"
fg_table = df[fg]

plt.figure(figsize = (12, 8))
plt.ylabel("Fitness Gadgets")
plt.setp(plt.bar(fg_table["Month-Year of Purchase"], fg_table["Units Sold"]), color = "blue", width = 0.4)

# Product Category = Travel Accessories
ta = df["Product Category"] == "Travel Accessories"
ta_table = df[ta]

plt.figure(figsize = (12, 8))
plt.ylabel("Travel Accessories")
plt.setp(plt.bar(ta_table["Month-Year of Purchase"], ta_table["Units Sold"]), color = "green", width = 0.4)

# Product Category = Flat Screen
fs = df["Product Category"] == "Flat Screen"
fs_table = df[fs]

# Concatenate product category tables into one table
tables = [dc_table, fg_table, ta_table, fs_table]
table = pd.concat(tables, keys = ["DC", "FG", "TA", "FS"])

display(table)
```
